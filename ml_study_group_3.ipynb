# **Import Library**

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

# **Cek File yang Sudah Diupload ke Google Colab**

```python
print("üìÇ Daftar File di Direktori /content/:")
print(os.listdir('/content/'))

file_path = "/content/student_scores.csv"

df = pd.read_csv(file_path)
```

# **Mengecek dan Menampilkan Statistik serta Isi Dataset**

```python
print("\nüîç Dataset Information:")
print(df.info())

print("\nüìä Ringkasan Statistik:")
print(df.describe())

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

print("\nüìú Student Scores Dataset:")
display(df)
```

# **Scatter Plot**

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['Hours'], y=df['Scores'], color='blue')
plt.title('Scatter Plot: Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

# **Regression Plot**

```python
plt.figure(figsize=(8,6))
sns.regplot(x=df['Hours'], y=df['Scores'], color='green', line_kws={"color": "red"})
plt.title('Regression Plot: Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

# **Heatmap Korelasi**

```python
plt.figure(figsize=(6,5))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", linewidths=0.5, fmt=".2f")
plt.title('Heatmap Korelasi')
plt.show()
```

## **Line Plot**

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=df['Hours'], y=df['Scores'], marker='o', color='purple')
plt.title('Line Plot: Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

# **Koefisien Korelasi**

```python
correlation = df.corr().iloc[0,1]
print(f"\n Koefisien Korelasi antara Variabel Hours dan Scores: {correlation:.2f}")
```

# **Insight hasil visualisasi data/EDA (Exploratory Data Analysis)**
Source dataset:
https://www.kaggle.com/datasets/samira1992/student-scores-simple-dataset


---


Dataset ini memiliki 25 entri untuk variabel *Hours* dan *Scores*. Tidak ditemukan data yang hilang (*missing values*), sehingga dataset ini lengkap untuk dianalisis. Rata-rata jumlah jam belajar adalah 5.01 jam, sementara rata-rata nilai adalah 51.48. Ini menunjukkan bahwa rata-rata siswa belajar sekitar 5 jam dan memperoleh nilai sekitar 51. Standar deviasi dari jumlah jam belajar adalah 2.53, dan untuk nilai adalah 25.29. Standar deviasi yang cukup besar ini menunjukkan adanya variasi dalam jumlah waktu belajar dan nilai yang diperoleh siswa.

Nilai minimum jam belajar adalah 1.1 jam, dan maksimum 9.2 jam, sementara nilai minimum adalah 17 dan maksimum 95. Siswa yang belajar paling sedikit (1.1 jam) memperoleh nilai terendah (17), sedangkan siswa yang belajar paling banyak (9.2 jam) memperoleh nilai tertinggi (95). Berdasarkan kuartil pertama (25%), 25% siswa belajar kurang dari 2.7 jam dan mendapatkan nilai kurang dari 30. Pada kuartil kedua (50%), separuh siswa belajar kurang dari 4.8 jam dan memperoleh nilai kurang dari 47. Sedangkan pada kuartil ketiga (75%), 25% siswa belajar lebih dari 7.4 jam dan mendapatkan nilai lebih dari 75.

Berdasarkan hasil visualisasi data, hubungan antara jumlah jam belajar (Hours) dan nilai yang diperoleh (Scores) terlihat positif. Scatter plot menunjukkan pola linear di mana semakin banyak waktu yang dihabiskan untuk belajar, semakin tinggi nilai yang diperoleh siswa. Data terdistribusi cukup rapi, meskipun ada sedikit variasi pada beberapa titik. Regresi plot menjelaskan hubungan ini dengan garis regresi yang menggambarkan tren peningkatan nilai seiring bertambahnya jam belajar. Garis tersebut memberikan gambaran bahwa hubungan antara jumlah jam belajar dan nilai dapat diwakili oleh sebuah model linier sederhana.

Visualisasi heatmap digunakan untuk menunjukkan kekuatan hubungan antar variabel dalam dataset, yaitu Hours dan Scores. Dengan anotasi korelasi yang ditampilkan di dalam heatmap, dapat dilihat seberapa kuat hubungan antara kedua variabel tersebut. Heatmap ini membantu memvisualisasikan hubungan antar variabel dengan cara yang intuitif, menggunakan gradasi warna untuk menunjukkan kekuatan hubungan.

Koefisien korelasi yang dihitung secara numerik menunjukkan nilai sebesar 0.98, yang menandakan adanya hubungan linier positif yang sangat kuat antara jumlah jam belajar dan nilai. Nilai ini menunjukkan bahwa peningkatan waktu belajar hampir sepenuhnya berkorelasi dengan peningkatan nilai siswa. Line plot juga mendukung hasil ini, menggambarkan kenaikan nilai yang konsisten seiring bertambahnya waktu belajar. Pola ini menunjukkan bahwa siswa yang belajar lebih lama cenderung memperoleh nilai yang lebih tinggi.

Berdasarkan analisis statistik dan visualisasi data, dapat disimpulkan bahwa waktu belajar merupakan faktor yang sangat signifikan dalam memengaruhi nilai siswa. Secara umum, siswa yang menghabiskan lebih banyak waktu untuk belajar cenderung mendapatkan nilai yang lebih tinggi. Nilai rata-rata meningkat seiring bertambahnya jumlah jam belajar, dan siswa yang belajar lebih dari 7.4 jam (di atas kuartil ketiga) umumnya memperoleh nilai yang sangat baik (di atas 75).

Visualisasi scatter plot dan regresi plot menunjukkan adanya hubungan linear yang kuat antara jumlah waktu belajar dan nilai siswa, dengan garis regresi yang menggambarkan tren peningkatan nilai secara konsisten. Selain itu, heatmap korelasi mendukung hubungan ini dengan memberikan representasi visual kekuatan hubungan antar variabel, sementara nilai koefisien korelasi sebesar 0.98 mengonfirmasi hubungan linier positif yang sangat kuat.

Secara keseluruhan, dapat disimpulkan bahwa data ini menunjukkan semakin lama waktu belajar dapat meningkatkan peluang siswa untuk meraih nilai yang lebih tinggi. Insight ini juga membuka peluang untuk menggunakan model prediktif berbasis regresi linier untuk memperkirakan nilai siswa berdasarkan waktu belajar yang dialokasikan.
